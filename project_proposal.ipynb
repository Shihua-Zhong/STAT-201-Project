```r
library(tidyverse)
library(digest)
library(repr)
library(tidymodels)
```

```r


temp <- tempfile()
download.file("https://archive.ics.uci.edu/ml/machine-learning-databases/00275/Bike-Sharing-Dataset.zip", temp)
data <- read_delim(unz(temp,"day.csv"),delim = ",", col_names = TRUE)
unlink(temp) 
data
```

```r
BikeShare <- data %>%
filter(yr == 0)%>%
filter(season %in% c(1,2)) %>%
select(season,weathersit)
```

Objective

We want to conduct a hypothesis test to determine if the proportion of bike rentals under a clear weather (weathersit = 1) between spring and winter are the same.


```r
sample_proportion <- BikeShare %>%
group_by(season) %>%
summarize(prop = sum(weathersit == 1)/n())
sample_proportion
```
